```
from statistics import mean
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style
style.use('ggplot')
import random
```

```
x=np.array([1,2,3,4,5,6],dtype=np.float64)
y=np.array([6,7,8,15,20,36],dtype=np.float64)
```

```
def create_dataset(length,variance,steps=2,correlation=False):
  val=1
  y=[]
  for i in range(length):
    z=val+random.randrange(-variance,variance)
    y.append(z)
    if correlation and correlation=='pos':
      val+=steps
    elif correlation and correlation=='neg':
      val-=steps
    
  x=[i for i in range(len(y))]
  return np.array(x,dtype=np.float64),np.array(y,dtype=np.float64)
```

```
x,y=create_dataset(100,100,2,correlation='pos')
```

```
print(y)
```

```
plt.scatter(x,y)
plt.show()
```

```
def best_fit_slope( x,y):
  m=((mean(x)*mean(y))-(mean (x*y)))/ ((mean(x)*mean(x))-mean(x*x))
  return m
```

```
m= best_fit_slope(x,y)
```

```
print(m)
```

```
def y_intercept(x,y,m):
  b=mean(y)-(m*mean(x))
  return b
```

```
b=y_intercept(x,y,m)
```

```
print( b)
```

```
regression_line=[((m*x)+b) for x in x]
```

```
style.use('fivethirtyeight')
```

```
predict_x=36
predict_y=(m*predict_x) + b
print(predict_y)
```

```
plt.plot(x,y,color='red',label='Data')
#plt.scatter(predict_x,predict_y,color='green', label='Predicted_data')
plt.plot(x,regression_line,color='Black',label='Regression_line')
plt.plot(predict_x,predict_y)
plt.legend(loc=0)
plt.show()
```

```
# r-square values needs to be as close as possible to 1
# r-square=1-(squared error for regression line/ squared error for y)
```

```
def squared_error(y,regression_line):
  return sum((regression_line-y)*(regression_line-y))
```

```
def coeff_of_regression(y,regression_line):
  squared_error_of_regression= squared_error(y,regression_line)
  
  y_mean=[mean(y) for i in y]
  squared_error_of_y=squared_error(y,y_mean)
  return 1-(squared_error_of_regression/squared_error_of_y)
```

```
def coefficient_of_determination(ys_orig,ys_line):
    y_mean_line = [mean(ys_orig) for y in ys_orig]

    squared_error_regr = sum((ys_line - ys_orig) * (ys_line - ys_orig))
    squared_error_y_mean = sum((y_mean_line - ys_orig) * (y_mean_line - ys_orig))

    print(squared_error_regr)
    print(squared_error_y_mean)

    r_squared = 1 - (squared_error_regr/squared_error_y_mean)

    return r_squared
```
